```python
import numpy as np 
import matplotlib
import matplotlib.pyplot as plt 
import array as arr
```

```python
N = 11 # of generation
t = range(N+1) #time 
x = np.zeros(N+1) #generation values
R = 2.5 #r-value
x[0] = 0.85 #initial generation value

n = range(N) #counting array 

for i in n: 
    x[i+1] = R*x[i]*(1-x[i])

figure1=plt.figure(figsize=(6,3),dpi=300)
numeric=plt.plot(t[0:N],x[0:N],'.',label='Numeric')
plt.title('Population vs Generation',fontsize = 16)
plt.xlabel('Generation', fontsize=14)
plt.xticks(fontsize=12)
plt.ylabel('Population', fontsize=14)
plt.yticks(fontsize=12)
plt.grid()
plt.axis([0, 11, 0,1.5])
figure1.savefig('generationvstime.pdf',bbox_inches='tight',dpi=300)
```
