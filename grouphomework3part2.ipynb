```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import array as arr
import math
```

```python
g_range=np.linspace(4,11,50000)
y_values=4
X = []
Y = []
for u in g_range:
    X.append(u)
    y_values = np.random.random()
    for n in range(1000):
      y_values=(u*y_values)*(1-math.tanh(y_values))
    Y.append(y_values)

plt.plot(X, Y, ls='', marker=',')
plt.show()
#noted divergence values are close to 5, 8, 9, 9.26ish, 9.35ish
```
